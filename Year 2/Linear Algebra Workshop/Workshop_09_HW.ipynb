Fitting a parabola:
Create a parabola(t,y) function which takes the input datapoints(t,y)and output the fitting formula and the plot

```python
import numpy as np
import matplotlib.pyplot as plt

def parabola(t, y):
    A = np.vstack([t ** 2, t, np.ones(len(t))]).T
    a, b, c = np.linalg.lstsq(A, y, rcond=None)[0]
    print(f"y = {c:.6f} {'+' if b >= 0 else ''} {b:.6f} t {'+' if a >= 0 else ''} {a:.6f} t^2")

    plt.plot(t, a * (t ** 2) + (b * t) + c)
    plt.plot(t, y, 'ro')
    

t=np.arange(10)
y0=20
v0=50

dy=np.random.randint(-50,50,10)/5
y=y0+v0*t-9.8/2*t**2+dy

print(t)
print(y)

parabola(t,y)
```
