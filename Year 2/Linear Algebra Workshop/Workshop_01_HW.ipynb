```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def color_mix(c1, c2):
    colour1 = np.array(c1)
    colour2 = np.array(c2)
    
    my_colour = colour1 + colour2
    
    for i in range (3):
        my_colour[i] /= 2
        
    return mcolors.to_rgb(my_colour)
```

```python
red = mcolors.to_rgb([1,0,0])
green = mcolors.to_rgb([0,1,0])
blue = mcolors.to_rgb([0,0,1])

colour = color_mix(red,blue)
plt.plot(2,4,'o', color = colour, markersize = 40)
plt.show()
```
